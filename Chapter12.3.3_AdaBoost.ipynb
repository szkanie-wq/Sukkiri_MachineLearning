# アダブーストを実装する

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

## データの読み込み

```python
df = pd.read_csv('Survived.csv')
df.head(3)
```

## 欠損値を穴埋めする

```python
def fill_age(df, pclass, survived, value):
    joken1 = df['Pclass'] == pclass
    joken2 = df['Survived'] == survived
    joken3 = df['Age'].isnull()
    df.loc[(joken1) & (joken2) & (joken3), 'Age'] = value
    return df
```

### ピボットテーブルで平均を調べる

```python
pt = df.pivot_table(index='Pclass', columns='Survived', values='Age')
print(pt)

for i in range(len(pt)-1):
    for j, val in enumerate(pt[i]):
        print(f'{i},{j},{val}')

```

### ピボットテーブルで調べた平均値で穴埋めする

```python
for i in range(len(pt)-1):
    for j, val in enumerate(pt[i]):
        print(f'{i},{j},{val}')
        df = fill_age(df, j+1, i, val)
```

## 文字データの列を数値に変換する

```python
# 特徴量として利用する列のリスト
col = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

x = df[col]
t = df['Survived']

# Sex列は文字の列なのでダミー変数化
dummy = pd.get_dummies(df['Sex'], drop_first=True, dtype=int)
x = pd.concat([x, dummy], axis=1)
x.head(3)
```

```python
# アダブーストのインポート
from sklearn.ensemble import AdaBoostClassifier
# ベースとなるモデル
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, t, test_size=0.2, random_state=0)

#　最大の深さ5の決定木を何個も作って行く
base_model = DecisionTreeClassifier(random_state=0, max_depth=5)
```

```python
# 決定木を500個作成
model = AdaBoostClassifier(n_estimators=500, random_state=0, estimator = base_model)
model.fit(x_train, y_train)
```

```python
print('訓練データの正解率：',model.score(x_train, y_train))
print('テストデータの正解率：',model.score(x_test, y_test))
```
