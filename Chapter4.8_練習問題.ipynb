```python
xcol = ['x0','x1','x2', 'x3']
x = df[xcol]
x.head(5)
```

```python
t = df['target']
t.head(5)
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier(random_state=0)

model.fit(x,t)
```

```python
x_anony = [['4.2','3.8','2.1','6.8']]

x_anony_df = pd.DataFrame(x_anony,columns=x.columns)

model.predict(x_anony_df)
```

```python
model.score(x,t)
```
