```python
xcol = ['身長','体重','年代']
x = df[xcol]
x.head(5)
```

```python
t = df['派閥']
t.head(5)
```

```python
from sklearn import tree
# モデルの準備(見学習)
model = tree.DecisionTreeClassifier(random_state = 0)
# 学習の実行
model.fit(x,t)
```

```python
# 一人についての派閥予測
taro = [[170, 70, 20]]
taro_df = pd.DataFrame(taro, columns=x.columns)

model.predict(taro_df)
```

```python
# 3人についての派閥予測
students = [[170, 70, 20], [172, 65, 20], [158,48,20]]
students_df = pd.DataFrame(students, columns=x.columns)

model.predict(students_df)
```

```python
#正解率の計算
model.score(x,t)
```

```python
# モデルの保存
import pickle

with open('KinokoTakenoko.pkl', 'wb') as f:
    pickle.dump(model,f)
```

```python
# 保存したモデルの読み込み

import pickle

with open('KinokoTakenoko.pkl', 'rb') as f:
    model2 = pickle.load(f)

# 他の3人についての派閥予測
new_students = [[158,48,20], [170, 70, 20], [172, 65, 20]]
new_students_df = pd.DataFrame(new_students, columns=x.columns)

model2.predict(new_students_df)
    
```
