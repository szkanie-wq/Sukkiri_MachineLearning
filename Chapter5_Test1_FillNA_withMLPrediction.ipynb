# がく片長さを予測する重回帰の予測モデルを作成する

```python
import pandas as pd
iris_df = pd.read_csv('iris.csv')
print(iris_df)
print(iris_df.isnull().sum())
```

```python
non_df=iris_df.dropna() # 欠損値を含む行を削除
print(non_df)
```

```python
from sklearn.linear_model import LinearRegression
x = non_df.loc[:, 'がく片幅':'花弁幅']
t = non_df['がく片長さ']
model = LinearRegression()
model.fit(x,t)
```

# 欠損データの本来の値を予測させる

```python
# 欠損行の抜き出し
condition = iris_df['がく片長さ'].isnull()
print(condition)
non_data = iris_df.loc[condition] # condition変数はTrueとFalseなので検索条件として利用できる
print(non_data)
```

```python
#　欠損行の入力に利用する特徴量だけを抜き出して、モデルで予測
x = non_data.loc[:, 'がく片幅':'花弁幅']
print(x)
pred = model.predict(x)
print(pred)
```

```python
# 欠損行のがく片長さのマスを抜き出して、predで代入
iris_df.loc[condition, 'がく片長さ'] = pred
```
