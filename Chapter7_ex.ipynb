```python
import pandas as pd
df = pd.read_csv('ex4.csv'); print(df.head(5))
columns = df.columns; print(columns)
print('Shape:',df.shape)
```

```python
# ユニークな値を調べる
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

for col in columns:
    if col != 'score':
        uniques = df[col].unique()
        sorted_uniques = np.sort(uniques)
        print(f'Unique item of {col} = {sorted_uniques}\n')
        value_counts = df[col].value_counts().sort_index()
        print(f'Each count of unique item of {col}:{value_counts}')

        value_counts.plot(kind='bar')
        plt.show()
```

```python
df['sex'].mean()
```

```python
df.groupby('class')['score'].plot(kind='hist',bins=10)
plt.xlabel('score')
plt.legend()
plt.title('Histgram of Score for each class')
plt.show()

mean_score = df.groupby('class')['score'].mean()
print(f'Mean Score of Group Class:',mean_score)
mean_score.plot(kind='bar')
plt.ylabel('Mean Score')
plt.show()
```

```python
# ピボットテーブルによる集計(平均値)
table = pd.pivot_table(df, index = 'sex', columns = 'class', values = 'score', aggfunc='mean'); print('Pivot Table\n',table)
```
